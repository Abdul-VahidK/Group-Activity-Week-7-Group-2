# 2-Visualisations of the Given Data

#### ●Region vs Surface Area & Agricultural production index

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))

plt.subplot(2,1,1)
sns.barplot(x='Region',y='Surface area (km2)',data=df)
plt.title('Rgion vs Surface Area',fontsize=15)
plt.ylabel('Surface Area')
plt.xticks(rotation=90)

plt.subplot(2,1,2)
sns.barplot(x='Region',y='Agricultural production index ',data=df)
plt.title('Region vs Agricultural production index',fontsize=15)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### Insights
    ●When compared to northern america, the surface area for south eastern asia is low.But even though their surface area is lower, they have the largest agricultural production index. 
    ●Regions like Micronesia, Polyesia, Caribbean have the least surface area, but they have Agricultural Production indices above 80. Means their production is increasing in the agricultural sector

#### ●Region vs Population & Imports

```python
plt.figure(figsize=(10,10))

plt.subplot(2,1,1)
sns.barplot(x='Region',y='Population in thousands (2017)',data=df)
plt.xticks(rotation=90)
plt.title('Region vs Population',fontsize=15)

plt.subplot(2,1,2)
sns.barplot(x='Region',y='International trade: Imports (million US$)',data=df)
plt.title('Region vs Imports',fontsize=15)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### Insights
    ●In Eastern Asia, there population is more, but they have relatively low surface area. It may be the reason for their high imports.
    ●Northern America has around 100K population,which is less than Eastern Asia and Southern Asia,they also have decent amount of surface area and API.But they have large imports,may be the imports are from non agricultural products category.

#### ●Region vs Number of Countries

```python
plt.figure(figsize=(9,6))
sns.countplot(df['Region'])
plt.xticks(rotation=90)
plt.title('Region vs Number of Countries',fontsize=16)
plt.ylabel('Number of Countries')
plt.show()
```

### Insights

# 4(a) -Region with Highest GDP Per capita

```python
df_gdp = df.groupby('Region')['GDP per capita (current US$)'].mean()
df_gdp.sort_values(ascending=False).head()
```

### North America has the highest GDP per capita with the value 57659 US dollars.

# 4(b) -Region with Highest International Trade: Imports

```python
df_import = df.groupby('Region')['International trade: Imports (million US$)'].mean()
df_import.sort_values(ascending=False).head()
```

### North America has the highest International trade: Imports with the value 663552.25 million US dollars.

# 5-Region Which is Consistent with Respect to GDP

```python
df_cons = df.groupby('Region')['GDP per capita (current US$)'].describe()
df_cons['CV'] = (df_cons['std']/df_cons['mean'])*100
df_cons.sort_values(by='CV',ascending=True).head()
```

### Oceania is most consistent region with respect to GDP
